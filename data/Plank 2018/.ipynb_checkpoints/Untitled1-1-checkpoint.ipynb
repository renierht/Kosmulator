```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as op
import emcee
import camb
import numpy as np
import healpy as hp
from multiprocess import Pool,get_context
import corner
```

```python
l = np.loadtxt('COM_PowerSpect_CMB-TT-binned_R3.01.txt', usecols = 0)
Dl = np.loadtxt('COM_PowerSpect_CMB-TT-binned_R3.01.txt', usecols = 1)
Dl_minus = np.loadtxt('COM_PowerSpect_CMB-TT-binned_R3.01.txt', usecols = 2)
Dl_plus = np.loadtxt('COM_PowerSpect_CMB-TT-binned_R3.01.txt', usecols = 3)
```

```python
plt.errorbar(l, Dl, yerr=[Dl_plus,Dl_minus], xerr=None, fmt='.')
plt.ylabel('D_l [μK^2]')
plt.xlabel('l')
plt.title('Power spectrum (TT)')
```

```python
def compute_power_spectrum(l, H0, Omega_m):
    params = camb.CAMBparams()
    params.set_cosmology(H0=H0, ombh2=0.0224, omch2=Omega_m, omk=0)
    params.InitPower.set_params(ns=0.965)
    params.set_for_lmax(int(l.max()), lens_potential_accuracy=1)
    results = camb.get_results(params)
    powers = results.get_cmb_power_spectra(params, CMB_unit='muK')
    Cl = powers['total'][:, 0]
    return Cl[:len(l)]
```

```python
def likelyhood(theta, l, Dl, Dl_plus, Dl_minus):
    Omega_m, H0 = theta
    model_Dl = compute_power_spectrum(l, H0, Omega_m)
    uncertainty = np.sqrt((Dl_plus - Dl) ** 2 + (Dl - Dl_minus) ** 2) / 2
    uncertainty = np.maximum(uncertainty, 1e-6)
    chi2 = np.sum(((Dl - model_Dl) / uncertainty) ** 2)
    return -0.5 * chi2
```

```python
def lnprior(theta):
    Omega_m,H0 = theta
    if 0 < Omega_m < 1 and 50 < H0 < 100:
        return 0.0
    return -np.inf
```

```python
def lnprob(theta,l,Dl,Dl_plus,Dl_minus):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + likelyhood(theta, l, Dl, Dl_plus, Dl_minus)
```

```python
H0 = 67.4
Omega_m0 = 0.27
```

```python
theta = [Omega_m0, H0]
nll = lambda *args: -likelyhood(*args)
Minimize = op.minimize(nll,theta,args=(l, Dl, Dl_plus, Dl_minus),bounds = [(0,0.4),(60,80)])
print (Minimize['x'])
```

```python
ndim = 2
nwalker = 5
nsteps = 10
burn = 5

pos = [theta + 1e-4 * np.random.randn(ndim) for _ in range(nwalker)]

with Pool() as pool:
    sampler = emcee.EnsembleSampler(nwalker,ndim,lnprob,args = (l,Dl,Dl_plus,Dl_minus), pool=pool)
    sampler.run_mcmc(pos,nsteps, progress=True)
    samples1 = sampler.chain[:,burn:,:].reshape((-1,ndim))
```

```python
fig = plt.figure()
corner.corner(samples1,labels=["omega_m","H"],color = "blue",plot_density=True, smooth=True,storechain=True, plot_datapoints=False, 
                        fill_contours=True,show_titles=True,title_fmt = '.3f',bins=100)
plt.savefig("Plank_2018.png",dpi=400)
plt.show()
```
